```python
%%time
lista = []
 
for valor in range(5):
    lista.append(valor + 10)

# print(lista)
```

```python
%%time
lista = [valor + 10 for valor in range(5)]
# print(lista)
```

```python
%%time
lista = []

for i in range(1, 30):
    if i % 4 == 0:
        lista.append(i)
```

```python
%%time
lista = [i for i in range(1, 30) if i % 4 == 0]
# print(lista)
```
